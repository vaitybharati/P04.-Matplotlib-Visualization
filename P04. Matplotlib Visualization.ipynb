# MatPlotLib Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from vega_datasets import data
```

```python
cars=data('cars')
```

## Table

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).head(10)
```

## Plotting two different categories

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).plot()
```

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).head(10).plot()
```

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).plot(kind='box')
```

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).head(10).plot(kind='bar')
```

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).plot(kind='hist')
```

```python
pd.crosstab(cars.Miles_per_Gallon,cars.Cylinders).head(10).plot(kind='hist')
```

## Plotting single category

```python
cars.Miles_per_Gallon.value_counts()
```

```python
f.ax = plt.subplots(figsize=(10,5))
cars.Miles_per_Gallon.value_counts().plot(kind='pie')
```

```python
cars.Miles_per_Gallon.head(10).value_counts().plot(kind='pie')
```

```python
cars.Miles_per_Gallon.head(10).value_counts().plot(kind='bar')
```

```python
cars.Name.value_counts()
```

```python

cars.Name.value_counts().plot(kind='pie')
```

```python
cars.Name.head(10).value_counts().plot(kind='pie')
```

## Scatter Plot

```python
plt.scatter(x='Name',y='Miles_per_Gallon',data=cars)
```

```python
plt.scatter(x='Name',y='Miles_per_Gallon',data=cars.head(3))
```

```python
plt.scatter(cars.Name,cars.Miles_per_Gallon)
```

```python
plt.scatter(cars.Name.head(5),cars.Miles_per_Gallon.head(5))
```

## Histogram

```python
plt.hist(cars.Cylinders)
```

```python
plt.hist(cars.Cylinders,facecolor='magenta',edgecolor='purple',bins=5)
```
